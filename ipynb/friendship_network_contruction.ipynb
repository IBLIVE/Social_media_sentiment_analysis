```python
import networkx as nx
import json

file_name = 'friendship_network_data.json'
with open(file_name, 'r') as f:
    friendship_network = json.load(f)

G = nx.Graph()
```

```python
for user_id, friends in friendship_network.items():
    user_id = int(user_id)
    G.add_node(user_id)     
    
for user_id, friends in friendship_network.items():
    user_id = int(user_id)
    for friend_id in friends:
        friend_id = int(friend_id) 
        G.add_edge(user_id, friend_id)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
for user_id, friends in friendship_network.items():
    user_id = int(user_id)
    G.add_node(user_id)     
    
    for friend_id in friends:
        friend_id = int(friend_id) 
        G.add_edge(user_id, friend_id)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
import matplotlib.pyplot as plt

nx.draw(G, with_labels=False, node_color='lightgreen', edge_color='gray', node_size=20)
plt.show()
```

```python
degrees = [G.degree(n) for n in G.nodes()]
```

```python
plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 1), color='lightgreen', edgecolor='black', log=False)
plt.title("Degree Distribution of a Network")
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 1), color='lightgreen', edgecolor='black', log=True)
plt.title("Degree Distribution of a Network (logarithmic)")
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.grid(True)
plt.show()
```

```python
clustering_data = nx.clustering(G)
closeness_data = nx.closeness_centrality(G)

clustering = [clustering_data[node] for node in G.nodes()]
closeness = [closeness_data[node] for node in G.nodes]
```

```python
plt.figure(figsize=(8, 6))
plt.hist(clustering, bins=30, color='lightblue', edgecolor='black', log=False)
plt.title("Clustering coefficients")
plt.xlabel("Coefficient")
plt.ylabel("Number of nodes")
plt.grid(True)
plt.show()
plt.figure(figsize=(8, 6))
plt.hist(clustering, bins=30, color='lightblue', edgecolor='black', log=True)
plt.title("Clustering coefficients (logarithmic)")
plt.xlabel("Coefficient")
plt.ylabel("Number of nodes")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(closeness, bins=10, color='lightcoral', edgecolor='black', log=False)
plt.title("Closeness centrality")
plt.xlabel("Centrality")
plt.ylabel("Number of nodes")
plt.grid(True)
plt.show()
```
